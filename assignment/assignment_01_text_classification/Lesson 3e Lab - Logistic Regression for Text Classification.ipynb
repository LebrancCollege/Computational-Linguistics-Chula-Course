```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split  
from sklearn.linear_model import LogisticRegression 
from sklearn.metrics import classification_report  
from nltk import word_tokenize
import nltk  
```

```python
nltk.download("punkt")
nltk.download("punkt_tab")
```

```python
train_data = pd.read_csv("../../data/csv/financial-news-data/financial-news-data/financial-news-train.csv")
test_data = pd.read_csv("../../data/csv/financial-news-data/financial-news-data/financial-news-test.csv")
```

```python
# Prepare Data 
train, train_dev = train_test_split(train_data, train_size=0.7, random_state=150)
test, test_dev = train_test_split(test_data, train_size=0.5, random_state=150)

# Feature Vector
vectorizor = CountVectorizer(tokenizer=nltk.word_tokenize)
train_feature_vectors = vectorizor.fit_transform(train["text"])

# Train Model
model = LogisticRegression(max_iter=1000)
model.fit(train_feature_vectors, train["label"]) 

# Evaluation
dev_feature_vector = vectorizor.transform(test_dev["text"])
dev_prediction = model.predict(dev_feature_vector)
print(classification_report(test_dev["label"], dev_prediction))   
```

```python
weights = pd.DataFrame(
  model.coef_,
  columns=vectorizor.get_feature_names_out(),
  index=model.classes_
)
```
