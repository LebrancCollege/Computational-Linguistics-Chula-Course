```python
import openai
import json
openai.api_key = ""
```

```python
def analyze_sentiment_with_reason(text):
    sentiment_analysis_prompt = f"""
    Analyze the sentiment of the text delimited by <text> and </text> tags.
    The sentiment must be one of the following: positive, negative, or neutral.

    Output must be json with keys:
        reason: the reason for the sentiment
        sentiment: the sentiment label

    <text>
    {text}
    </text>
    """
    messages = [
            {"role": "user", "content": sentiment_analysis_prompt}
        ]
    response = openai.chat.completions.create(
        model="gpt-4o-mini",
        messages=messages,
        response_format={ "type": "json_object" }
    )
    json_response = json.loads(response.choices[0].message.content)
    return json_response

sentiment = analyze_sentiment_with_reason("บริษัทจะต้องเน้นการสร้างความเชื่อมั่นให้กับลูกค้าให้มากขึ้น")
```

```python
def analyze_sentiment(text):
    sentiment_analysis_prompt = f"""
    Analyze the sentiment of the text delimited by <text> and </text> tags.
    The sentiment must be one of the following: positive, negative, or neutral.

    Output must be label string only

    <text>
    {text}
    </text>
    """
    messages = [
            {"role": "user", "content": sentiment_analysis_prompt}
        ]
    response = openai.chat.completions.create(
        model="gpt-4o-mini",
        messages=messages,
    )
    return response.choices[0].message.content

sentiment_label = analyze_sentiment("บริษัทจะต้องเน้นการสร้างความเชื่อมั่นให้กับลูกค้าให้มากขึ้น")

```
